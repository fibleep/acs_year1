```python
x: list[int] = [4, 5, 6, 7, 8, 9, 10]
frequency: list[int] = [2, 3, 5, 6, 4, 2, 3]
x_frequency: DataFrame = pd.DataFrame({'Frequency': frequency}, index=x)
x_frequency
```

```python
# Cumulative sum percentages THE CORRECT WAY
((x_frequency.Frequency / x_frequency.Frequency.sum()).sort_index().cumsum() * 100).round(1)
```

```python
# Percentages
((x_frequency.Frequency / x_frequency.Frequency.sum()).sort_index() * 100).round(1)
```
